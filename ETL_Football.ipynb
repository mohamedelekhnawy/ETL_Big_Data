# ETL in python using Web Scraping 

```python
from bs4 import BeautifulSoup
import requests
```

## <li>Extract data from the website</li>

```python
url = 'https://en.wikipedia.org/wiki/List_of_professional_sports_leagues_by_revenue'
page = requests.get(url)
soup = BeautifulSoup(page.text, 'html')
```

```python
print(soup)
```

```python
soup.find('table')
```

```python
soup.find_all('table')[0]
```

```python
soup.find('table', class_ = 'wikitable sortable')
```

```python
table = soup.find_all('table')[0]
```

```python
print(table)
```

```python
world_titles = table.find_all('th')
```

```python
world_titles
```

```python
world_table_titles = [title.text.strip() for title in world_titles]

print(world_table_titles)
```

## <li>Transfrom Data into dataframe</li>

```python
import pandas as pd
```

```python
df = pd.DataFrame(columns = world_table_titles)

df
```

```python
column_data = table.find_all('tr')
```

```python
for row in column_data[1:]:
    row_data = row.find_all('td')
    individual_row_data = [data.text.strip() for data in row_data]
    
    length = len(df)
    df.loc[length] = individual_row_data
```

## <li>Load the dataframe into PostgreSQL database</li>

```python
import psycopg2
import pandas as pd
from sqlalchemy import create_engine

# Define your DataFrame with the specified columns
data = {
    'Rank': list(range(1, 101)),
    'League': ['Premier League'] * 100,
    'Sport': ['Football'] * 100,
    'Country(ies)': ['England'] * 100,
    'Season': [2021] * 100,
    'Tier Level': [1] * 100,
    'Teams[a]': [20] * 100,
    'Mat­ches / Games / Events': [380] * 100,
    'Rev.(€ mil)': [5000] * 100,
    'Rev. / team(€ mil)': [250] * 100,
    'Rev. / match(€ thou­sands)': [13] * 100,
    'Ref': ['Link1'] * 100
}

df = pd.DataFrame(data)

# Connect to your PostgreSQL database
conn = psycopg2.connect(
    host="youer_host",
    database="youer_database",
    user="youer_user",
    password="youer_pass"
)

# Create a cursor object
cur = conn.cursor()

# Define your PostgreSQL table creation query
create_table_query = '''CREATE TABLE IF NOT EXISTS  Leagues (
    Rank INTEGER,
    League VARCHAR(100),
    Sport VARCHAR(100),
    Country VARCHAR(100),
    Season INTEGER,
    Tier_Level INTEGER,
    Teams INTEGER,
    Matches INTEGER,
    Revenue_mil INTEGER,
    Revenue_per_team_mil INTEGER,
    Revenue_per_match_thousands INTEGER,
    Reference VARCHAR(255)
);'''

# Execute the table creation query
cur.execute(create_table_query)

# Commit changes to the database
conn.commit()

# Create a connection engine using SQLAlchemy
engine = create_engine('postgresql://youer_user:root@youer_host:5432/ "youer_database",
')

# Load the DataFrame into the PostgreSQL database
df.to_sql('your_table_name', engine, if_exists='replace', index=False)

# Close the cursor and connection
# cur.close()
# conn.close()
```
